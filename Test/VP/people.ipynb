```python
# Доля пропусков в каждом столбце
missing_data = data.isnull().mean() * 100
print(missing_data)

# Столбцы с большим количеством пропусков (например, больше 50%)
train = missing_data[missing_data > 50].index
print("Столбцы с более чем 50% пропусков:")
print(train)

```

```python
# Проверка на дубликаты столбцов
duplicates = data.T.duplicated(keep='first')  # Ищем дублирующиеся столбцы
duplicate_columns = data.columns[duplicates]

print("Дублирующиеся столбцы:")
print(duplicate_columns)
```
