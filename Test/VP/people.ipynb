Читаем наш датасет с помощью библиотеки pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
import numpy as np

# Загружаем данные из pandas
df = pd.read_csv('../csv_file/valid.csv')
df.head(10)
```

```python
print("Valid Data Info ----------------------------------------------:")
print(df.info())
print("\nValid Data Head -------------------------------------------:")
print(df.head())
```

```python
df.isnull().sum()
```

```python
df = df.replace([np.inf, -np.inf, np.nan], 0)
df.drop_duplicates()
df.shape
```

```python
print("Размер данных до удаления:", df.shape)
data = df.dropna(axis=1, how='all')
print("Размер данных после удаления:", data.shape)
```

```python
missing_percentage = data.isnull().mean() * 100

print("Процент пропусков в каждой колонке:")
print(missing_percentage)
```

```python
sns.countplot(x='target', data=data)
plt.title('Распределение целевой переменной')
plt.show()
```

```python
cols = data.columns
#фиолетовые - не пропущенные данные, желтые - пропущенные
colours = ['#8b00ff', '#ffff00']
sns.heatmap(data[cols].isnull(), cmap=sns.color_palette(colours))
```
