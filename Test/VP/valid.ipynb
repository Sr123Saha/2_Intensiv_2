```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
import numpy as np

# Загружаем данные из pandas
df = pd.read_csv('../csv_file/valid.csv')
df.head(10)
```

```python
df.info()

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates(subset=["client_id"]) 
```

```python
print("Размер данных после удаления:", df.shape)
```

```python
missing_percentage = df.isnull().mean() * 100

print("Процент пропусков в каждой колонке:")
print(missing_percentage)
```

```python
missing_data = 2400
missing_counts = df.isnull().sum()
columns_to_drop = missing_counts[missing_counts > missing_data].index
data = df.drop(columns=columns_to_drop)
```

```python
print("Размер данных после удаления:", data.shape)
```

```python
missing_percentage = data.isnull().mean() * 100

print("Процент пропусков в каждой колонке:")
print(missing_percentage)
```

```python
sns.countplot(x='target', data=data)
plt.title('Распределение целевой переменной')
plt.show()
```
