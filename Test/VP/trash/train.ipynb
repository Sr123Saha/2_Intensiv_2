```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
import numpy as np
```

```python
# Загружаем данные из pandas
data = pd.read_csv('../csv_file/train.csv')
```

Читаем данные из файла и понимаем, что очень много пропусков

```python
data.info()
```

Смотрит колличество пропусков в столбце и замечаем что 4 столбика не имеют пропусков.

```python
data.isnull().sum()
```

Отбрасываем столбцы в которых очень много пропусков (больше 90%) и удаляем дубликаты по столбцу "Client_ID" чтобы не было одинаковых клиентов.

```python
threshold = 0.9
print(len(data))
deleate = int(len(data) * (1 - threshold))
df_clean = data.dropna(axis=1,thresh=deleate)
df_clean = df_clean.drop_duplicates(subset=['client_id'])
df_clean
```

Смотрим колличество столбиков и строк

```python
df_clean.shape
```

```python
cols = df_clean.columns
#фиолетовые - не пропущенные данные, желтые - пропущенные
colours = ['#8b00ff', '#ffff00']
sns.heatmap(df_clean[cols].isnull(), cmap=sns.color_palette(colours))
```
