```python
import matplotlib.pyplot as plt

dv = 65
ac = 72
pv = 76 
main = 78

models = ['baseline Долич', 'baseline Серый', 'baseline Подъячев', 'общая модель']
accuracies = [dv, ac, pv, main]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['orange', 'green', 'blue', 'red'])

plt.title('Сравнение точности моделей')
plt.xlabel('Модели')
plt.ylabel('Точность (%)')

plt.show()
```
